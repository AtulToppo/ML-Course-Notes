# Scipy Library
SciPy is a scientific computation library that uses NumPy underneath.
SciPy stands for Scientific Python.
It provides more utility functions for optimization, stats and signal processing.
Like NumPy, SciPy is open source so we can use it freely.
SciPy was created by NumPy's creator Travis Olliphant.

```python
import scipy
```

```python
#For a particular module (here, linear algebra)

import scipy.linalg as la
import numpy as np
a = np.array([[1,2],[3,4]])
a
```

```python
#Determinant

la.det(a)
```

```python
#Works on lists also

import scipy.linalg as la
import numpy as np
a = ([[1,2],[3,4]])
la.det(a)
```

```python
#Inverse

import scipy.linalg as la
import numpy as np
a = ([[1,2],[3,4]])
la.inv(a)
```

```python
#Inverse of singular ,atrices

import scipy.linalg as la
import numpy as np
a = ([[1,2],[2,4]])
la.inv(a)
```

```python
#Matrix Multiplication

a = ([[1,2],[3,4]])
b=[[5],[6]]
np.dot(a,b)
```

```python
#Solving system of linear equation using inverse

a = ([[1,2],[3,4]])
b=[[5],[6]]
la.inv(a).dot(b)
```

```python
#Other Way

a = ([[1,2],[3,4]])
b=[[5],[6]]
la.solve(a,b)
```

```python
#Eigen vector

a = ([[1,2],[3,4]])
b=[[5],[6]]
la.eig(a)
```

```python
#svd = Singular value decomposition

a = ([[1,2],[3,4]])
b=[[5],[6]]
la.svd(a)
```

```python
#LU = lowerâ€“upper decomposition

a = ([[1,2],[3,4]])
b=[[5],[6]]
la.lu(a)
```

```python
import scipy.interpolate as si
import numpy as np
x=np.linspace(0,1,20)
noise=(np.random.random(20)*2-1)*1e-1    
y=np.sin(2*np.pi*x)+noise
cubicmodel=si.interp1d(x,y,kind='cubic')

x1=np.linspace(0,1,10)
cubicresult=cubicmodel(x1)

import matplotlib.pyplot as plt
plt.plot(x,y)
plt.plot(x1,cubicresult,'ro')
```

```python
import scipy.optimize as sp
x=np.linspace(-10,10,200)
def f(x):
    return x**2
sp.minimize(f,x0=-5)
```

```python
plt.plot(x,f(x))
```

```python
import scipy.optimize as sp
x=np.linspace(-10,10,200)
def f(x):
    return x**2+10*np.sin(x)
sp.minimize(f,x0=-5)
res=sp.minimize(f,x0=-5)
res
```

```python
plt.plot(x,f(x))
plt.plot(res['x'],res['fun'],'o')
plt.axhline(0)
```

```python
import scipy.optimize as sp
x=np.linspace(-10,10,200)
def f(x):
    return x**2+10*np.sin(x)
sp.minimize(f,x0=-5)
res=sp.minimize(f,x0=2,method='BFGS')
res

plt.plot(x,f(x))
plt.plot(res['x'],res['fun'],'o')
plt.axhline(0)
```
