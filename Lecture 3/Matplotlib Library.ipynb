# Matplotlib Library
Matplotlib is a low level graph plotting library in python that serves as a visualization utility. It was created by John D. Hunter.
Matplotlib is mostly written in python, a few segments are written in C, Objective-C and Javascript for Platform compatibility.

```python
import matplotlib as mp
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

The plot() function is used to draw points (markers) in a diagram.
By default, the plot() function draws a line from point to point.

The function takes parameters for specifying points in the diagram.
Parameter 1 is an array containing the points on the x-axis.
Parameter 2 is an array containing the points on the y-axis.

```python
x=np.linspace(0,10,100)
y=np.sin(x)
fig=plt.figure()
ax=plt.axes()
ax.plot(x,y)
```

# Scatter Plots
A scatter plot uses dots to represent values for two different numeric variables. The position of each dot on the horizontal and vertical axis indicates values for an individual data point. Scatter plots are used to observe relationships between variables.

```python
x=np.linspace(0,10,50)
plt.scatter(x,np.sin(x))
```

##### rand = random values
##### randn = normally distributed random data (most of the points will be around 0)

```python
#To produce the symbols in different sizes

x=np.linspace(0,10,50)
size=np.random.rand(50)
plt.scatter(x,np.sin(x),s=size)
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
size=300*np.random.rand(50)
plt.scatter(x,y,s=size)
```

```python
#To produce the symbols in different colors

x=np.random.rand(50)
y=np.random.rand(50)
size=300*np.random.rand(50)
colors=np.random.rand(50)
plt.scatter(x,y,s=size,c=colors)
```

```python
#To produce the colorbar

x=np.random.rand(50)
y=np.random.rand(50)
size=300*np.random.rand(50)
colors=np.random.rand(50)
plt.scatter(x,y,s=size,c=colors)
plt.colorbar()
colors
```

```python
#cmap (Color Mapping) for a specific colour

x=np.random.rand(50)
y=np.random.rand(50)
size=300*np.random.rand(50)
colors=np.random.rand(50)
plt.scatter(x,y,s=size,c=colors,cmap='Blues')
plt.colorbar()
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
size=300*np.random.rand(50)
colors=np.random.rand(50)
plt.scatter(x,y,s=size,c=colors,cmap='Greens')
plt.colorbar()
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
size=300*np.random.rand(50)
colors=np.random.rand(50)
plt.scatter(x,y,s=size,c=colors,cmap='Reds')
plt.colorbar()
```

# Pie Chart

```python
Z = np.random.uniform(0,1,8)
plt.pie(Z)
plt.show()
```

```python
# Labels

Z = [10,20,40,30,5]
label = ['apple','banana','cherry','guava','plum']
plt.pie(Z, labels = label)
plt.show()
```

```python
# Start Angle

Z = [10,20,40,30,5]
label = ['apple','banana','cherry','guava','plum']
plt.pie(Z, labels = label, startangle=180)
plt.show()
```

```python
# Explode

Z = [10,20,40,30,5]
label = ['apple','banana','cherry','guava','plum']
explode=[0.1,0,0,0,0]
plt.pie(Z, labels = label, startangle=180, explode=explode)
plt.show()
```

```python
# Shadow

Z = [10,20,40,30,5]
label = ['apple','banana','cherry','guava','plum']
explode=[0.1,0,0.5,0,0]
plt.pie(Z, labels = label, startangle=180, explode=explode, shadow=True)
plt.show()
```

```python
# Colors

Z = [10,20,40,30,5]
label = ['apple','banana','cherry','guava','plum']
color = ['red','orange','yellow','green','pink']
explode=[0.1,0,0.5,0,0]
plt.pie(Z, labels = label, colors = color,startangle=180, explode=explode,shadow=True)
plt.show()
```

```python
# Legend

Z = [10,20,40,30,5]
label = ['apple','banana','cherry','guava','plum']
color = ['red','orange','yellow','green','pink']
explode=[0.1,0,0.5,0,0]
plt.pie(Z, labels = label, colors = color,startangle=180, explode=explode,shadow=True)
plt.legend()
plt.show()
```
